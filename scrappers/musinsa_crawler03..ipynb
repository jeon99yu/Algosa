```python
import requests
import pandas as pd

data_list = []

for page in range(1, 2):
    url = f'https://api.musinsa.com/api2/dp/v1/plp/goods?gf=A&category=104005001&size=60&caller=CATEGORY&page={page}'
    res = requests.get(
        url,
        headers={
            "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                          "AppleWebKit/537.36 (KHTML, like Gecko) "
                          "Chrome/138.0.0.0 Safari/537.36"
        }
    )

    if res.status_code != 200:
        print(f"[Error] {page}페이지 요청 실패")
        continue

    items = res.json().get("data", {}).get("list", [])
    for item in items:
        data_list.append({
            "product_id": item.get("goodsNo"),
            "brandName": item.get("brandName"),
            "goodsName": item.get("goodsName"),
            "price": item.get("price"),
            "reviewCount": item.get("reviewCount"),
            "reviewScore": item.get("reviewScore"),
            "thumbnail": item.get("thumbnail"),
            "goodsLinkUrl": item.get("goodsLinkUrl"),
        })

df = pd.DataFrame(data_list)
df.to_csv("data/musinsa_products.csv", index=False, encoding="utf-8-sig")

print(df.head(10))
```
